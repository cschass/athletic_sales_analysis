```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
Athletic_sales_2020_df= pd.read_csv('Resources/athletic_sales_2020.csv')
Athletic_sales_2021_df= pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
Athletic_sales_2020_df.head(20)
```

```python
# Display the 2021 sales DataFrame
Athletic_sales_2021_df.head(20)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
Athletic_sales_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
Athletic_sales_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_df = pd.concat([Athletic_sales_2020_df, Athletic_sales_2021_df],axis="rows", join="inner")
combined_sales_df
```

```python
# Check if any values are null.
combined_sales_df.isnull().sum()
```

```python
# Check the data type of each column
combined_sales_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_df['invoice_date'] = pd.to_datetime(combined_sales_df['invoice_date'], format='%Y-%m-%d')
combined_sales_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
#Show the number products sold for region, state, and city.                    
# Rename the sum to "Total_Products_Sold".

product_by_location_df = combined_sales_df.groupby(['region', 'state', 'city'])["total_sales"].sum().reset_index(name='Total_Products_Sold')

# Show the top 5 results.
product_by_location_df.head(5)

```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
product_by_location_df =pd.pivot_table(combined_sales_df,
                                                values='units_sold',
                                                index=['region', 'state', 'city'],
                                                aggfunc='sum').reset_index()
product_by_location_df = product_by_location_df.rename(columns={"units_sold" : "Total_Products_Sold"})
product_by_location_df.head(5)

# Show the top 5 results.
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

product_sales_by_location_df = combined_sales_df.groupby(['region', 'state', 'city'])["total_sales"].sum().reset_index(name='Total_Sales')

# Show the top 5 results.
product_sales_by_location_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
# Optional: Rename the "total_sales" column to "Total Sales"

product_sales_by_location_df =pd.pivot_table(combined_sales_df,
                                                values='total_sales',
                                                index=['region', 'state', 'city'],
                                                aggfunc='sum').reset_index()
product_sales_by_location_df = product_sales_by_location_df.rename(columns={"total_sales" : "Total_Sold"})

# Show the top 5 results.
product_sales_by_location_df.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"
invoice__df = pd.pivot_table(combined_sales_df,
                                                values='total_sales',
                                                index=['invoice_date'],
                                                aggfunc='sum').reset_index()
invoice__df = invoice__df.rename(columns={"total_sales" : "Total Sales"})
invoice__df.head(5)

# Show the table.
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

daily_sales = invoice__df.set_index('invoice_date').resample('D').sum().reset_index()
# Sort the resampled pivot table in descending order on "Total Sales".

daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
daily_sales.head(5)

```
